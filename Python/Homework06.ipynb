## Problem 1

Write a function called `group_means` that takes its input a Pandas DataFrame with exactly the columns: "labels" and "weights". Group entries by labels and return the average weights for each label. Make sure to name the column in the returned data frame "average weight" and use the "labels" as the index, but don't name the index column. 

**Example:** See an example by executing the code in the second cell below.

```python
import pandas as pd
import numpy as np
from IPython.display import display

def group_means(X):
    new = pd.DataFrame(X.groupby('labels').mean()) #Groups them all with like labels and finds the mean
    new.rename(columns = {'weights' : 'average weight'}, inplace = True) #Renames the weights column
    new.index.name = None #Removed the lables label from the now index
    return(new)
    
```

- Include only the countries in the region "Europe & Central Asia"
- Use the "Country Name" as the row identifier or Index value.
- Include the following columns for each country:
    * population count in 2000 
    * population count in 2020 
    * percent growth in population in \% between 2000 and 2020 (this should be a negative number if the population in 2020 is smaller than in 2000). Example: if population grows from 100k to 105k, then the growth is 5\%.
    * income group of the country.
    
Display your table in Data Frame format in your notebook so that the grader can see it. 

```python
# YOUR CODE HERE

#Reads in files and creates data frames
metaDataFile = "/Users/zanderbonnet/Desktop/Math 167PS/Homework/Data/WorldBank_Metadata_Countries.csv"
metaData = pd.read_csv(metaDataFile)
popDataFile = "/Users/zanderbonnet/Desktop/Math 167PS/Homework/Data/WorldBank_Population.csv"
popData = pd.read_csv(popDataFile)


#Gets the country codes of all the European and Central Asian countries
metaData.set_index(["Region"], inplace = True)
metaDataTrim = metaData.loc[("Europe & Central Asia")]
metaDataTrim.reset_index(inplace = True)
codes = metaDataTrim["Country Code"]


#Collects the populations of the countries for both 2000 and 2020 with matching country codes to the meta data
popData.set_index(["Country Code"], inplace = True)
pop2000 = []
pop2020 = []
for i in codes:
    if(i in popData.index):
        pop2000.append(popData.loc[(i)]['2000'])
        pop2020.append(popData.loc[(i)]['2020'])
        

countName = metaDataTrim[['Country Name']] #Gets the country names
pop2000s = pd.Series(pop2000) 
pop2020s = pd.Series(pop2020) #Creates a series so that it can be added to data frame
popGrow = round(((pop2020s - pop2000s)/pop2000s)*100,3) #Gets teh growth percentage
popGrows = pd.Series(popGrow)
 #Creates a data frame of values to add from the population dataset
toAdd = pd.DataFrame({"Population in 2000":pop2000s, "Population in 2020":pop2020s, "Population Growth(%)": popGrows})
incomeGroup = metaDataTrim[['Income Group']]
comb = pd.concat([countName,toAdd, incomeGroup], axis = 1) #Combines all the parts of the data set
comb.set_index(['Country Name'], inplace = True) #Sets index of the data set to the country names
comb
```
